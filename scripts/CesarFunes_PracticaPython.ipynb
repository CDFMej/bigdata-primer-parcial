# Ejercicio 1: Clasificación de edad

**Enunciado:** determinar si una persona es niño, adolescente, adulto o adulto mayor según su edad.

**Logica:** se evalua la edad con condiciones if, elif y else . El primer rango que se cumple determina la
categoría.

```python
def clasificar_edad(edad):
    if edad < 13:
        return "Niño"
    elif edad <= 17:
        return "Adolescente"
    elif edad <= 59:
        return "Adulto"
    else:
        return "Adulto mayor"

edad = 3
categoria = clasificar_edad(edad)
print(f"Edad: {edad} anos")
print(f"Categoria: {categoria}")
```

# Ejercicio 2: Conversor de temperatura
**Enunciado:** Crear una funcion para convertir grados celsius a farenheit.

**Logica:** Se obtiene los grados en celsius para convertirlo en farenheit se multiplica por 9/5 y al resultado de esto se le suman 32 y devuelve la cantidad en farenheit.


```python
def convertir_celsius_fahrenheit(c):
    return (c * 9/5) + 32

celsius = 30
fahrenheit = convertir_celsius_fahrenheit(celsius)
print(f"Celsius: {celsius}")
print(f"Fahrenheit: {fahrenheit}")
```

# Ejercicio 3: Calculadora basica
**Enunciado:** Crear una función para calcular operaciones matematicas basicas multiplicación, división, suma y resta.

**Logica:** Se mandan los 2 numeros a la función y la operación a realizar, con secuencias de if, elif y else se hace la operación basandose en la operación mandada, tambien hay una validacio´n mediante un if para validar que no se pueda dividir entre 0.

```python
def calculadora(a, b, operacion):
    if operacion == "suma":
        return a + b
    elif operacion == "resta":
        return a - b
    elif operacion == "multiplicacion":
        return a * b
    elif operacion == "division":
        if b == 0:
            return "Error: no se puede dividir entre cero"
        return a / b

num1 = 10
num2 = 2
op = "division"

resultado = calculadora(num1, num2, op)
print(f"Operacion: {op}")
print(f"Resultado: {resultado}")
```

**Enunciado:** Crear una función para definir si un numero es par o impar.

**Logica:** Se manda un numero a la función que hace una división entre 2 para ver si el resto es 0 el numero es par y si no es impar.

# Ejercicio 9: Mayor y menor de una lista

**Enunciado:** Crear una funcion que saque el numero maximo y minimo de una lista y calcular la diferencia que hay entre ellos.

**Logica:** Se utliza la función max que recorre el arreglo y devuelve el valor maximo del arreglo y la función min que hace lo mismo con el valor minimo, estos valores se asignaron a una variable y se restaron para conseguir la diferencia que hay entre ellos.

```python
def mayor_o_menor(lista):
    maximo= max(lista)
    minimo= min(lista)
    diferencia= maximo - minimo
    print(f"El numero Maximo es: {maximo}")
    print(f"El numero Minimo es: {minimo}")
    print(f"La diferencia entre estos es de: {diferencia}")

listas = [1,5,15,20,100,8,9,25,62]

mayor_o_menor(listas)
    
```

# Ejercicio 10:  Diccionario de estudiante 

**Enunciado:** Crear una funcion que muestre los datos de un dicccionario en formato ordenado.

**Logica:** Para acceder a los valores de un dicionario se hace mediante clave **por ejemplo estudiante["edad"] = 21**.

```python
def dicciordenado(estudiante):
    
    diccionario_ordenado = dict(sorted(estudiante.items()))
    
    print("--- Datos del Estudiante ---")
    for clave, valor in diccionario_ordenado.items():
        print(f"{clave.capitalize()}: {valor}")

estudiante = {
    "nombre": "Carlos Mendoza",
    "edad": 21,
    "carrera": "Ingeniería en Sistemas",
    "promedio": 88.5,
    "estado": "Activo"
}

dicciordenado(estudiante)
```

# Ejercicio 11: Funcion para calcular descuento

**Enunciado:** Crear una función que se ingrese el valor final y el descuento para devolver el valor con el descuento aplicado.

**Logica:** Los valores que entran en la función son los de el precio original y el de descuento aplicado que se divide entre 100, luego se multiplica por el precio original y despues se resta el resultado al precio original dandonos así el precio final que es el que retorna la función.

```python
def calcular_descuento(precio_original, porcentaje_descuento):
    descuento = precio_original * (porcentaje_descuento / 100)
    precio_final = precio_original - descuento
    return precio_final

precio_producto = 1500
descuento_aplicado = 15

resultado = calcular_descuento(precio_producto, descuento_aplicado)
print(f"El precio final con el {descuento_aplicado}% de descuento es: Lps. {resultado}")
```

# Ejercicio 12:  Validación de contraseña 

**Enunciado:** Crear una función que valide que una contraseña cumpla con ciertos los requisitos.

**Logica:** Se envia la contraseña a la función, se valida mediante **len** que no tenga menos de 8 caracteres, la de **isdigit** para validar que contenga al menos un numero, la de **isupper** para validar que contenga al menos una mayúscula y la de **islower** para validar que contenga al menos una minúscula, en caso de no cumplir con alguna de estas validaciones devuelve false y un mensaje con el motivo por el cual la contraseña no es valida.

```python
def validar_contrasena(password):
    if len(password) < 8:
        return False, "La contraseña debe tener al menos 8 caracteres."
    
    tiene_numero = False
    tiene_mayuscula = False
    tiene_minuscula = False
    
    for caracter in password:
        if caracter.isdigit():
            tiene_numero = True
        elif caracter.isupper():
            tiene_mayuscula = True
        elif caracter.islower():
            tiene_minuscula = True
            
    if not tiene_numero:
        return False, "La contraseña debe contener al menos un número."
    if not tiene_mayuscula:
        return False, "La contraseña debe contener al menos una letra mayúscula."
    if not tiene_minuscula:
        return False, "La contraseña debe contener al menos una letra minúscula."
        

    return True, "¡Contraseña válida y segura!"

clave_usuario = input("Defina su nueva contraseña: ")
es_valida, mensaje = validar_contrasena(clave_usuario)

if es_valida:
    print(f" {mensaje}")
else:
    print(f" Contraseña INVÁLIDA. Motivo: {mensaje}")
```

# Ejercicio 13: Sistema simple de inventario

**Enunciado:** Crear una lista de productos con código, nombre, precio y cantidad disponible. Mostrar todos los productos, buscar uno por código y calcular el valor total del inventario.

**Lógica:** Se usa una lista de diccionarios para guardar los productos. Luego se recorren con un ciclo for para mostrarlos, buscar un código y calcular el total.

```python
productos = [
    {"codigo": 101, "nombre": "Laptop", "precio": 15000, "cantidad": 3},
    {"codigo": 102, "nombre": "Mouse", "precio": 500, "cantidad": 10},
    {"codigo": 103, "nombre": "Teclado", "precio": 1200, "cantidad": 5}
]

print("Lista de productos:")

for producto in productos:
    print(producto)

codigo_buscar = 102

for producto in productos:
    if producto["codigo"] == codigo_buscar:
        print("\nProducto encontrado:")
        print(producto)

total_inventario = 0

for producto in productos:
    total_producto = producto["precio"] * producto["cantidad"]
    total_inventario += total_producto

print("\nValor total del inventario:", total_inventario)
```

# Ejercicio 14: Registro de ventas

## Enunciado
Crear una lista de ventas donde cada venta tenga cliente, producto, cantidad y precio unitario.
Calcular el total por venta, el total general vendido y el cliente con la compra más alta.

## Lógica utilizada
Se utiliza una lista de diccionarios para almacenar las ventas.
Se recorren los datos con un ciclo for para calcular los totales y encontrar la compra más alta.


```python
ventas = [
    {"cliente": "Juan", "producto": "Laptop", "cantidad": 1, "precio": 15000},
    {"cliente": "Ana", "producto": "Mouse", "cantidad": 2, "precio": 500},
    {"cliente": "Luis", "producto": "Teclado", "cantidad": 3, "precio": 1200}
]

print("Registro de ventas:\n")

total_general = 0
mayor_compra = 0
cliente_mayor = ""

for venta in ventas:
    total_venta = venta["cantidad"] * venta["precio"]

    print("Cliente:", venta["cliente"])
    print("Producto:", venta["producto"])
    print("Total venta:", total_venta)
    print("-------------------")

    total_general += total_venta

    if total_venta > mayor_compra:
        mayor_compra = total_venta
        cliente_mayor = venta["cliente"]

print("\nTotal general vendido:", total_general)
print("Cliente con la compra más alta:", cliente_mayor)
```

## Explicación
Se usó una lista de diccionarios para guardar las ventas.
El ciclo for permitió recorrer cada venta y calcular el total multiplicando cantidad por precio.
También se utilizó una condición if para identificar la compra más alta.

# Ejercicio 15: Limpieza básica de texto

## Enunciado
Definir una frase, eliminar espacios innecesarios, convertir el texto a minúsculas,
reemplazar espacios dobles y contar la cantidad de palabras.

## Lógica utilizada
Se aplican métodos de texto para limpiar y organizar la frase.
Luego se divide el texto en palabras para contarlas.

```python
texto = "   Hola    Mundo Python   "

print("Texto original:")
print(texto)

texto = texto.strip()

texto = texto.lower()

texto = texto.replace("  ", " ")

palabras = texto.split()

cantidad_palabras = len(palabras)

print("\nTexto limpio:")
print(texto)

print("\nCantidad de palabras:", cantidad_palabras)
```

## Explicación
El método strip() elimina espacios al inicio y final del texto.
El método lower() convierte todas las letras a minúsculas.
El método replace() reemplaza espacios dobles por un solo espacio.
El método split() divide el texto en palabras usando espacios.
Finalmente, len() cuenta la cantidad de palabras.
