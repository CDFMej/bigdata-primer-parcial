# Ejercicio 16: DataFrame de estudiantes
* Cree un DataFrame con al menos 10 estudiantes.
* Incluya nombre, carrera, edad, promedio y ciudad.
* Muestre las primeras 5 filas.
* Muestre la información general del DataFrame.
* Calcule el promedio general de calificaciones.
* Muestre estudiantes con promedio mayor o igual a 80.
* Ordene estudiantes por promedio de mayor a menor.


```python
import pandas as pd

def crear_dataframe():
    datos = {
        "nombre": ["Ana", "Luis", "Carlos", "María", "Sofía", "Pedro", "Lucía", "Jorge", "Valeria", "Diego"],
        "carrera": ["Ingeniería", "Medicina", "Derecho", "Arquitectura", "Ingeniería", "Economía", "Medicina", "Derecho", "Arquitectura", "Economía"],
        "edad": [20, 22, 21, 23, 19, 24, 22, 21, 23, 20],
        "promedio": [85, 78, 90, 88, 95, 70, 82, 76, 91, 84],
        "ciudad": ["Tegucigalpa", "San Pedro Sula", "La Ceiba", "Comayagua", "Choluteca", "El Progreso", "Santa Rosa", "Tela", "Danlí", "Siguatepeque"]
    }

    df = pd.DataFrame(datos)
    return df

df_estudiantes = crear_dataframe()

print("Primeras 5 filas:")
print(df_estudiantes.head())

print("\nInformación general del DataFrame:")
print(df_estudiantes.info())

promedio_general = df_estudiantes["promedio"].mean()
print("\nPromedio general de calificaciones:", promedio_general)

print("\nEstudiantes con promedio mayor o igual a 80:")
print(df_estudiantes[df_estudiantes["promedio"] >= 80])

print("\nEstudiantes ordenados por promedio de mayor a menor:")
print(df_estudiantes.sort_values(by="promedio", ascending=False))
```

# Ejercicio 17: Análisis de ventas con pandas
* Cree un DataFrame de ventas con al menos 15 registros.
* Incluya fecha, vendedor, producto, categoría, cantidad y precio unitario.
* Calcule una nueva columna TotalVenta.
* Calcule el total vendido por vendedor.
* Identifique el producto más vendido.
* Identifique la categoría con mayor ingreso.
* Ordene las ventas de mayor a menor.


```python
import pandas as pd

def crear_dataframe_ventas():
    datos = {
        "fecha": [
            "2026-01-01","2026-01-02","2026-01-03","2026-01-04","2026-01-05",
            "2026-01-06","2026-01-07","2026-01-08","2026-01-09","2026-01-10",
            "2026-01-11","2026-01-12","2026-01-13","2026-01-14","2026-01-15"
        ],
        "vendedor": [
            "Ana","Luis","Carlos","Ana","Sofía",
            "Pedro","Luis","Carlos","Ana","Sofía",
            "Pedro","Luis","Carlos","Ana","Sofía"
        ],
        "producto": [
            "Laptop","Mouse","Teclado","Monitor","Laptop",
            "Mouse","Teclado","Monitor","Laptop","Mouse",
            "Teclado","Monitor","Laptop","Mouse","Teclado"
        ],
        "categoria": [
            "Tecnología","Accesorios","Accesorios","Tecnología","Tecnología",
            "Accesorios","Accesorios","Tecnología","Tecnología","Accesorios",
            "Accesorios","Tecnología","Tecnología","Accesorios","Accesorios"
        ],
        "cantidad": [2,5,3,1,4,6,2,1,3,7,4,2,1,5,3],
        "precio_unitario": [800,20,30,200,800,20,30,200,800,20,30,200,800,20,30]
    }

    df = pd.DataFrame(datos)
    return df

df_ventas = crear_dataframe_ventas()

df_ventas["TotalVenta"] = df_ventas["cantidad"] * df_ventas["precio_unitario"]

print("DataFrame de ventas:")
print(df_ventas)

print("\nTotal vendido por vendedor:")
print(df_ventas.groupby("vendedor")["TotalVenta"].sum())

print("\nProducto más vendido:")
print(df_ventas.groupby("producto")["cantidad"].sum().sort_values(ascending=False).head(1))

print("\nCategoría con mayor ingreso:")
print(df_ventas.groupby("categoria")["TotalVenta"].sum().sort_values(ascending=False).head(1))

print("\nVentas ordenadas de mayor a menor:")
print(df_ventas.sort_values(by="TotalVenta", ascending=False))
```

 # Ejercicio 18:
 En este ejercicio se hace una limpieza de datos esencial en un DataFrame con registros de clientes incompletos. Primero, normaliza la columna de ciudades convirtiendo los espacios en blanco ocultos en nulos reales (NaN) para luego identificar y contar de forma global cuántas ausencias existen por columna mediante isnull().sum(). Posteriormente, se corrigen los datos faltantes usando dos estrategias distintas: rellena los huecos de la columna ciudad con el texto "No especificada" a través de fillna(), y elimina únicamente las filas que carecen de un dato crítico como el correo utilizando dropna(subset=['correo']), logrando así una tabla depurada y lista para su análisis sin perder información valiosa de otras columnas.

```python
import pandas as pd
import numpy as np

data = {
    'nombre': ['Ana Gómez', 'Carlos Pérez', 'María López', 'Luis Martínez', 'Elena Rostova'],
    'edad': [25, np.nan, 30, 22, np.nan],
    'correo': ['ana@email.com', 'carlos@email.com', None, 'luis@email.com', 'elena@email.com'],
    'ciudad': ['Madrid', 'Barcelona', 'Sevilla', None, '    '],  # Incluye un espacio en blanco
    'telefono': ['600111222', None, '600333444', '600444555', None]
}

df = pd.DataFrame(data)
print("--- dataframe original incompleto ---")
print(df)
print("\n" + "="*50 + "\n")

df['ciudad'] = df['ciudad'].replace(r'^\s*$', np.nan, regex=True)
valores_nulos = df.isnull()

print("--- cantidad de valores nulos---")
print(df.isnull().sum())
print("\n" + "="*50 + "\n")

df['ciudad'] = df['ciudad'].fillna("No especificada")

df = df.dropna(subset=['correo'])

print("--- dataframe limpio ---")
print(df)
```

# Ejercicio 19 - Análisis de préstamos

```python
import pandas as pd

# Crear datos de solicitudes de préstamo
datos_prestamos = {
    "Cliente": [
        "Carlos", "María", "José", "Ana", "Luis",
        "Sofía", "Pedro", "Elena", "Miguel", "Laura"
    ],
    "MontoSolicitado": [
        45000, 75000, 30000, 90000, 50000,
        120000, 25000, 65000, 40000, 85000
    ],
    "PlazoMeses": [
        12, 24, 6, 36, 18,
        48, 12, 24, 6, 36
    ],
    "IngresoMensual": [
        18000, 32000, 15000, 45000, 22000,
        60000, 14000, 28000, 17000, 38000
    ],
    "EstadoSolicitud": [
        "Aprobado", "Aprobado", "Rechazado", "Aprobado", "Pendiente",
        "Aprobado", "Rechazado", "Pendiente", "Rechazado", "Aprobado"
    ]
}

# Convertir el diccionario en un DataFrame
prestamos = pd.DataFrame(datos_prestamos)

print("DataFrame de solicitudes de préstamo:")
display(prestamos)

# Calcular monto promedio solicitado
monto_promedio = prestamos["MontoSolicitado"].mean()
print("Monto promedio solicitado:", monto_promedio)

# Contar solicitudes por estado
solicitudes_por_estado = prestamos["EstadoSolicitud"].value_counts()
print("\nSolicitudes por estado:")
print(solicitudes_por_estado)

# Mostrar únicamente préstamos aprobados
prestamos_aprobados = prestamos[prestamos["EstadoSolicitud"] == "Aprobado"]
print("\nPréstamos aprobados:")
display(prestamos_aprobados)

# Mostrar clientes con monto solicitado mayor a 50,000
clientes_mayor_50000 = prestamos[prestamos["MontoSolicitado"] > 50000]
print("\nClientes con monto solicitado mayor a 50,000:")
display(clientes_mayor_50000)

# Agrupar por estado y calcular promedio del monto solicitado
promedio_por_estado = prestamos.groupby("EstadoSolicitud")["MontoSolicitado"].mean()
print("\nPromedio del monto solicitado por estado:")
print(promedio_por_estado)
```

En este ejercicio se creó un DataFrame llamado prestamos usando un diccionario llamado datos_prestamos. 
Cada clave del diccionario representa una columna, por ejemplo Cliente, MontoSolicitado, PlazoMeses, IngresoMensual y EstadoSolicitud.

La variable prestamos guarda la tabla completa creada con pd.DataFrame(). 
La variable monto_promedio guarda el promedio de la columna MontoSolicitado usando el método mean(). 
La variable solicitudes_por_estado cuenta cuántas solicitudes hay por cada estado usando value_counts(). 
La variable prestamos_aprobados filtra únicamente las filas donde EstadoSolicitud es igual a "Aprobado". 
La variable clientes_mayor_50000 filtra los clientes cuyo MontoSolicitado es mayor a 50000. 
Finalmente, groupby() agrupa los datos por EstadoSolicitud y mean() calcula el promedio del monto solicitado para cada grupo.

# Ejercicio 20: Mini análisis de datos tipo empresa 

**Enunciado:** Consiste en realizar un análisis descriptivo básico sobre una estructura de datos empresariales mediante la librería pandas, con el objetivo de extraer métricas clave de los empleados, segmentar la información por departamentos, filtrar bajo criterios específicos de experiencia y clasificar al personal en rangos profesionales.

**Logica:** El programa opera de forma secuencial cargando primero los datos en un objeto DataFrame para habilitar el procesamiento en bloque; posteriormente, utiliza funciones estadísticas nativas para los cálculos generales, aplica agrupaciones de filas (groupby) para los promedios indexados, genera máscaras booleanas para filtrar filas que cumplan condiciones matemáticas y segmenta variables continuas en categorías discretas mediante la función pd.cut.

```python
import pandas as pd


datos = {
    'Nombre': ['Ana Gómez', 'Luis Martínez', 'María Rodríguez', 'Carlos Pérez', 'Sofía Ruiz', 'Juan Soria', 'Elena Vega'],
    'Departamento': ['IT', 'FTE', 'IT', 'Ventas', 'FTE', 'Ventas', 'IT'],
    'Cargo': ['Desarrollador', 'Analista', 'Tech Lead', 'Ejecutivo', 'Gerente', 'Asistente', 'Data Scientist'],
    'Salario': [45000, 38000, 75000, 32000, 80000, 25000, 62000],
    'Años_Experiencia': [3, 4, 9, 2, 12, 1, 6],
    'Ciudad': ['Madrid', 'Barcelona', 'Madrid', 'Sevilla', 'Valencia', 'Barcelona', 'Madrid']
}

df = pd.DataFrame(datos)


print("--- Primeros Registros del DataFrame ---")
print(df.head(3))  
print("\n" + "-"*50 + "\n")


print("--- Estadísticas Generales de Salarios ---")
print(df['Salario'].describe())
print("\n" + "-"*50 + "\n")


print("--- Salario Promedio por Departamento ---")
salario_promedio = df.groupby('Departamento')['Salario'].mean()
print(salario_promedio)
print("\n" + "-"*50 + "\n")


print("--- Empleado con el Salario Más Alto ---")
empleado_top = df.loc[df['Salario'].idxmax()]
print(f"Nombre: {empleado_top['Nombre']} | Departamento: {empleado_top['Departamento']} | Salario: ${empleado_top['Salario']}")
print("\n" + "-"*50 + "\n")


print("--- Empleados con más de 5 años de experiencia ---")
expertos = df[df['Años_Experiencia'] > 5]
print(expertos[['Nombre', 'Años_Experiencia', 'Cargo']])
print("\n" + "-"*50 + "\n")


print("--- Empleados ordenados por salario (Descendente) ---")
df_ordenado = df.sort_values(by='Salario', ascending=False)
print(df_ordenado[['Nombre', 'Salario']])
print("\n" + "-"*50 + "\n")


rangos = [-1, 2, 5, float('inf')] 
etiquetas = ['Junior', 'Intermedio', 'Senior']

df['NivelExperiencia'] = pd.cut(df['Años_Experiencia'], bins=rangos, labels=etiquetas)

print("--- DataFrame Final con la columna NivelExperiencia ---")
print(df)
```
