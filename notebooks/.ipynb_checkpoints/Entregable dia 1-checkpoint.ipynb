# Entregable Día 1 : introducción al Big Data
#### Empresa Netflix

# ¿Qué datos genera?
R/ 
- **Tipo de datos:** Datos de comportamiento como el tiempo en el que miras las series, las series que no terminas de ver, el dispositivo en el que mas usas la plataforma, metadatos de videos como el genero, actores entre otros.
- **Volumen:** Gernera vairos petabytes de datos al día debido a la cantidad de clientes con los que cuenta.
- **Frecuencia:** La frecuencia sería de las 24 horas del día debido a que es una plataforma de streaming

# ¿Quíen los usa?
R/
- **Equipos de ingeniería de datos y algoritmos:** Para mejorar el motor utilizado para la recomendación de contenido.
- **Directores de Contenido y Productores:** Para decidir que tipo series o peliculas financiar basandose en los gustos de la audiencia. 
- **Equipos de infraestructura de red:** Para predecir qué servidores locales deben almacenar temporalmente un video antes de que la gente lo pida.

# ¿Qué problema resuelve?
R/
Con el análisis de estos datos se resuelven diferentes problemas, por ejemplo: la retención de los usuraios mediante la recomendación de un contenido afín a este, la eficiencia de costes mediante la inversion en contenido que puede llegar a gustar a la mayoría o un amplio numero de usuarios.

# ¿Por qué no bastaría Excel?
R/ Basandonos en las 4 Vs:
- **Volumen:** Excel empieza a fallar a partir de un millon de registros, Netflix supera esta cantidad en eventos en algunos segundos.
- **Velocidad:** Las recomendaciones deben cambiar en milisegundos mientras el usuario navega por la app, excel no procesa flujo de datos en tiempo real.
- **Variedad:** Netflix cruza datos como fechas e Id de usuarios con Logs de servidores y texto de busqueda que excel no puede correlacionar eficientemente.
- **Veracidad/Complejidad:** Los algoritmos de Machine Learning que calculan la probabilidad de que veas una pelicula o serie requieren clústeres de servidores distribuidos, algo que nos e puede hacer con excel.


# Justificación de por qué es Big Data
El sistema de Netflix es un caso de Big Data por que cumple con las 4 Vs(Volumen, Velocidad, Variedad y Veracidad/Complejidad) la cantidad de datos generados y la variedad que tiene no puede ser manejada por un solo ordenador por lo cual usan una arquitectura en la nube para transformar las millones de interaciones en una interfaz de usuario personalizada.
